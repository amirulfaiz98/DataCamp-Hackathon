```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('census5.csv')
```

```python
income0 = df[df.income == 0]
income1 = df[df.income == 1]
plt.scatter(income1['hours-per-week'],income1['income'], label='above 50K', alpha=0.05)
plt.scatter(income0['hours-per-week'],income0['income'],label='below 50K', alpha= 0.05)
plt.xlabel('Working Hour per Week')
plt.ylabel('Income')
plt.title('Q1')
plt.legend()
plt.show()
```

```python
x1 = df[df['education-num']=='A']
x2 = df[df['education-num']=='C']

plt.subplot(1,2,1)
plt.hist(x1['income'], color='blue')
plt.title('Education A')
plt.xlabel('Income')

plt.subplot(1,2,2)
plt.hist(x2['income'], color='red')
plt.title('Education C')
plt.xlabel('Income')

plt.tight_layout()
plt.show()
```

```python
z = pd.DataFrame(df['capital-gain']-df['capital-loss'])
print(z)

plt.scatter(z,df['income'],alpha=0.1)
plt.xlabel('Investment')
plt.ylabel('Income')
plt.title('People with high income level has a lot of investment')
plt.show()
```
