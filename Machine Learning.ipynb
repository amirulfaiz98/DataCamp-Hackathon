```python
import pandas as pd
```

```python
from IPython.display import Markdown, display
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
import numpy as np
from sklearn import metrics    
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn import tree
```

```python
df = pd.read_csv('census5.csv')
```

```python
df.head()
```

```python
adult_data = df.drop(columns='income')
adult_label = df.income
```

```python
adult_data.head()
```

```python
adult_cat_1hot = pd.get_dummies(adult_data.select_dtypes('object'))
```

```python
adult_cat_1hot.head()
```

```python
adult_non_cat = adult_data.select_dtypes(exclude = 'object')
adult_data_1hot = pd.concat([adult_non_cat, adult_cat_1hot], axis=1, join='inner')
```

```python
# Train - Test split 
X_train, X_test, y_train, y_test = train_test_split(adult_data_1hot, adult_label, test_size  = 0.20,random_state=42, stratify=adult_label)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_test, y_test))
```

```python
print(knn.score(X_train, y_train))
```
